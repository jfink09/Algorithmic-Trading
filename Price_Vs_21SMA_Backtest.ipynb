```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

from backtesting.test import SMA, GOOG

class SmaCross(Strategy):
    def init(self):
        price = self.data.Close
        self.ma = self.I(SMA, price, 21)

    def next(self):
        price = self.data.Close
        if crossover(price, self.ma):
            self.buy()
        elif crossover(self.ma, price):
            self.sell()
```

```python
bt = Backtest(GOOG, SmaCross, commission=0, cash=10000,
              exclusive_orders=True)
stats = bt.run()
bt.plot()
```
